## 3. Обработка запросов списком команд (p, s, l, ads, ds, q)


````python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def doc_info(document_list,directories_dict):
    doc_number = input("Введите номер документа: ")
    for doc in document_list:
        if doc_number == doc['number']:
            return f"Владелец документа: {doc['name']}" 
    else:
         return 'Документ не найден в базе'   
        
def doc_info_1(document_list,directories_dict):
    doc_number = input("Введите номер документа: ") 
    for shelf, list_doc in directories_dict.items():
        if doc_number in list_doc:
            return f"Документ хранится на полке: {shelf}" 
    else:
         return 'Документ не найден в базе'    

        
def doc_info_2(document_list,directories_dict):
    for info in document_list:
        doc_number = info['number']
        for shelf, list_doc in directories_dict.items():
            if doc_number in list_doc:
                print(f"№: {info['number']} тип: {info['type']} владедец: {info['name']} полка хранения: {shelf}")


def change_info(document_list,directories_dict):
    shelf_number = input('Введите номер полки:')
    var = ''
    if shelf_number in directories_dict.keys():
        var = "Такая полка уже существует. Текущий перечень полок:" + ','.join(list(directories_dict))
    else:
        directories[shelf_number] = []
        var = 'Полка добавлена. Текущий перечень полок:' + ','.join(list(directories_dict))
    return var
                  
                     
def change_info1(document_list,directories_dict):
    shelf_number = input('Введите номер полки:')
    var = ' '
    if shelf_number not in directories_dict.keys():
        var = 'Такой полки не существует. Текущий перечень полок:' + ','.join(list(directories_dict))
    elif len(directories_dict[shelf_number]) == 0:
        del directories_dict[shelf_number]
        var = 'Полка удалена. Текущий перечень полок:' + ','.join(list(directories_dict))       
    else:
        var = 'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:' + ','.join(list(directories_dict))
    return var
                 
        
def main(document_list, directories_dict):
    command_dict = {
        'p': doc_info,
        's': doc_info_1,
        'l': doc_info_2,
        'ads': change_info,
        'ds': change_info1 
    }
  
    while True:
        command_request = input("Введите команду: ")
        if command_request in command_dict:
            res = command_dict[command_request](documents,directories)         
            print(res)
        elif command_request == "q":
            print('Работа закончена.')
            break
main(documents, directories)


````
